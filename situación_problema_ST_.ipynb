```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.tsa.stattools import adfuller, kpss, coint
from scipy import stats
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.ar_model import AutoReg
from statsmodels.tsa.arima_process import ArmaProcess
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import os
from IPython.display import display, HTML
import warnings

# Suppress warnings
warnings.filterwarnings('ignore')

class TimeSeriesAnalyzer:
    """
    A class for analyzing financial time series data.

    This class provides methods for:
    - Loading data from Excel files
    - Checking stationarity
    - Differencing time series as needed
    - Fitting AR, ARMA, and ARIMA models
    - Forecasting future values
    - Testing for cointegration between series
    - Visualizing results
    """

    def __init__(self):
        """Initialize the TimeSeriesAnalyzer."""
        self.data = {}  # Dictionary to store dataframes for each stock
        self.stationary_data = {}  # Dictionary to store stationary series
        self.differencing_order = {}  # Dictionary to store differencing orders
        self.models = {}  # Dictionary to store fitted models
        self.best_models = {}  # Dictionary to store best models
        self.forecasts = {}  # Dictionary to store forecasts

    def load_data(self, file_path, stock_name=None):
        """
        Load data from an Excel file.

        Parameters:
        ----------
        file_path : str
            Path to the Excel file
        stock_name : str, optional
            Name of the stock. If None, the file name without extension is used.

        Returns:
        -------
        pd.DataFrame
            The loaded dataframe
        """
        if stock_name is None:
            stock_name = os.path.splitext(os.path.basename(file_path))[0]

        # Load data
        df = pd.read_excel(file_path)

        # Ensure there's a 'Close' column
        if 'Close' not in df.columns:
            raise ValueError(f"No 'Close' column found in {file_path}")

        # Convert the first column to datetime if it's not already
        if not pd.api.types.is_datetime64_any_dtype(df.iloc[:, 0]):
            try:
                df.iloc[:, 0] = pd.to_datetime(df.iloc[:, 0])
            except:
                print(f"Warning: Could not convert first column to datetime in {file_path}")

        # Set the first column as index
        df = df.set_index(df.columns[0])

        # Store the dataframe
        self.data[stock_name] = df

        print(f"Loaded data for {stock_name}: {len(df)} rows")
        return df

    def check_stationarity(self, stock_name):
        """
        Check stationarity of a time series using ADF and KPSS tests.

        Parameters:
        ----------
        stock_name : str
            Name of the stock

        Returns:
        -------
        tuple
            (is_stationary, adf_results, kpss_results)
        """
        if stock_name not in self.data:
            raise ValueError(f"No data found for {stock_name}")

        series = self.data[stock_name]['Close']

        # ADF test
        adf_result = adfuller(series, autolag='AIC')
        adf_stationary = adf_result[1] < 0.05

        # KPSS test
        kpss_result = kpss(series, regression='c', nlags='auto')
        kpss_stationary = kpss_result[1] > 0.05

        # Series is considered stationary if both tests agree
        is_stationary = adf_stationary and kpss_stationary

        print(f"\nStationarity Test Results for {stock_name}:")
        print(f"ADF Test - p-value: {adf_result[1]:.4f} ({'Stationary' if adf_stationary else 'Non-stationary'})")
        print(f"KPSS Test - p-value: {kpss_result[1]:.4f} ({'Stationary' if kpss_stationary else 'Non-stationary'})")
        print(f"Overall: {'Stationary' if is_stationary else 'Non-stationary'}")

        return is_stationary, adf_result, kpss_result

    def difference_series(self, stock_name, max_diff=2):
        """
        Difference a time series until it becomes stationary.

        Parameters:
        ----------
        stock_name : str
            Name of the stock
        max_diff : int, optional
            Maximum number of differencing operations

        Returns:
        -------
        pd.Series
            The differenced series
        """
        if stock_name not in self.data:
            raise ValueError(f"No data found for {stock_name}")

        series = self.data[stock_name]['Close']

        # Check if already stationary
        is_stationary, _, _ = self.check_stationarity(stock_name)

        if is_stationary:
            print(f"Series for {stock_name} is already stationary. No differencing needed.")
            self.stationary_data[stock_name] = series
            self.differencing_order[stock_name] = 0
            return series

        # Apply differencing
        diff_order = 0
        diff_series = series.copy()

        for i in range(1, max_diff + 1):
            diff_series = diff_series.diff().dropna()
            diff_order = i

            # Create a temporary dataframe to check stationarity
            temp_df = pd.DataFrame({'Close': diff_series})
            temp_stock_name = f"{stock_name}_temp"
            self.data[temp_stock_name] = temp_df

            is_stationary, _, _ = self.check_stationarity(temp_stock_name)

            # Remove the temporary dataframe
            self.data.pop(temp_stock_name)

            if is_stationary:
                print(f"Series became stationary after {diff_order} differencing operations.")
                break

        if not is_stationary:
            print(f"Warning: Series did not become stationary after {max_diff} differencing operations.")

        self.stationary_data[stock_name] = diff_series
        self.differencing_order[stock_name] = diff_order

        return diff_series

    def fit_models(self, stock_name, max_order=3):
        """
        Fit AR, ARMA, and ARIMA models to the time series.

        Parameters:
        ----------
        stock_name : str
            Name of the stock
        max_order : int, optional
            Maximum order for AR, MA terms

        Returns:
        -------
        dict
            Dictionary of fitted models
        """
        if stock_name not in self.stationary_data:
            # First ensure we have a stationary series
            self.difference_series(stock_name)

        diff_series = self.stationary_data[stock_name]
        d = self.differencing_order[stock_name]

        models = {}
        model_results = {}

        # Determine the best order for each model type using AIC
        best_ar_order = 1
        best_ar_aic = float('inf')
        best_arma_orders = (1, 1)
        best_arma_aic = float('inf')
        best_arima_orders = (1, d, 1)
        best_arima_aic = float('inf')

        print(f"\nFitting models for {stock_name}:")

        # Try different AR orders
        for p in range(1, max_order + 1):
            try:
                model = AutoReg(diff_series, lags=p)
                result = model.fit()
                aic = result.aic

                if aic < best_ar_aic:
                    best_ar_aic = aic
                    best_ar_order = p
                    models[f'AR({p})'] = model
                    model_results[f'AR({p})'] = result
            except:
                print(f"Failed to fit AR({p}) model")

        # Try different ARMA orders
        for p in range(max_order + 1):
            for q in range(max_order + 1):
                if p == 0 and q == 0:
                    continue
                try:
                    model = ARIMA(diff_series, order=(p, 0, q))
                    result = model.fit()
                    aic = result.aic

                    if aic < best_arma_aic:
                        best_arma_aic = aic
                        best_arma_orders = (p, q)
                        models[f'ARMA({p},{q})'] = model
                        model_results[f'ARMA({p},{q})'] = result
                except:
                    print(f"Failed to fit ARMA({p},{q}) model")

        # Try different ARIMA orders
        for p in range(max_order + 1):
            for q in range(max_order + 1):
                if p == 0 and q == 0:
                    continue
                try:
                    # For ARIMA, we use the original series and specify the d parameter
                    model = ARIMA(self.data[stock_name]['Close'], order=(p, d, q))
                    result = model.fit()
                    aic = result.aic

                    if aic < best_arima_aic:
                        best_arima_aic = aic
                        best_arima_orders = (p, d, q)
                        models[f'ARIMA({p},{d},{q})'] = model
                        model_results[f'ARIMA({p},{d},{q})'] = result
                except:
                    print(f"Failed to fit ARIMA({p},{d},{q}) model")

        # Print best models for each type
        print(f"Best AR model: AR({best_ar_order}) with AIC = {best_ar_aic:.2f}")
        print(f"Best ARMA model: ARMA{best_arma_orders} with AIC = {best_arma_aic:.2f}")
        print(f"Best ARIMA model: ARIMA{best_arima_orders} with AIC = {best_arima_aic:.2f}")

        # Determine the overall best model
        best_model_name = min(model_results, key=lambda x: model_results[x].aic)
        best_aic = model_results[best_model_name].aic

        print(f"Overall best model for {stock_name}: {best_model_name} with AIC = {best_aic:.2f}")

        self.models[stock_name] = model_results
        self.best_models[stock_name] = best_model_name

        return model_results

    def forecast(self, stock_name, periods=30, confidence=0.95):
        """
        Forecast future values using the best fitted model.

        Parameters:
        ----------
        stock_name : str
            Name of the stock
        periods : int, optional
            Number of periods to forecast
        confidence : float, optional
            Confidence level for prediction intervals

        Returns:
        -------
        pd.DataFrame
            Forecasted values with confidence intervals
        """
        if stock_name not in self.models:
            self.fit_models(stock_name)

        best_model_name = self.best_models[stock_name]
        model_result = self.models[stock_name][best_model_name]

        # Get the last date in the original series
        last_date = self.data[stock_name].index[-1]

        # Create a date range for the forecast
        if isinstance(last_date, pd.Timestamp):
            # If we have a datetime index, create a date range
            forecast_dates = pd.date_range(start=last_date, periods=periods + 1)[1:]
        else:
            # If we have a numeric index, create a range
            forecast_dates = range(len(self.data[stock_name]) + 1, len(self.data[stock_name]) + periods + 1)

        # Get the forecasted values and confidence intervals based on model type
        if 'AR' in best_model_name and 'ARMA' not in best_model_name and 'ARIMA' not in best_model_name:
            # For AutoReg models
            forecast_mean = model_result.predict(start=len(self.data[stock_name]), end=len(self.data[stock_name]) + periods - 1)

            # Get forecast variance (this is model specific)
            forecast_var = np.ones_like(forecast_mean) * model_result.sigma2

            # Create confidence intervals manually
            from scipy import stats
            alpha = 1 - confidence
            crit_value = stats.norm.ppf(1 - alpha/2)

            lower_ci = forecast_mean - crit_value * np.sqrt(forecast_var)
            upper_ci = forecast_mean + crit_value * np.sqrt(forecast_var)

        else:
            # For ARMA/ARIMA models that have get_forecast method
            try:
                # Get forecast
                forecast = model_result.get_forecast(steps=periods)
                forecast_mean = forecast.predicted_mean

                # Get confidence intervals
                forecast_ci = forecast.conf_int(alpha=1-confidence)
                lower_ci = forecast_ci.iloc[:, 0]
                upper_ci = forecast_ci.iloc[:, 1]
            except:
                # Fallback method if get_forecast fails
                forecast_mean = model_result.forecast(steps=periods)

                # Create simple confidence intervals (less accurate but works as fallback)
                forecast_var = np.ones_like(forecast_mean) * model_result.sigma2
                crit_value = stats.norm.ppf(1 - (1-confidence)/2)

                if isinstance(forecast_mean, pd.Series):
                    lower_ci = forecast_mean - crit_value * np.sqrt(forecast_var)
                    upper_ci = forecast_mean + crit_value * np.sqrt(forecast_var)
                else:
                    lower_ci = pd.Series(forecast_mean - crit_value * np.sqrt(forecast_var))
                    upper_ci = pd.Series(forecast_mean + crit_value * np.sqrt(forecast_var))

        # Create a DataFrame with the forecast
        forecast_df = pd.DataFrame({
            'Forecast': forecast_mean,
            'Lower CI': lower_ci,
            'Upper CI': upper_ci
        }, index=forecast_dates)

        print(f"\nForecast for {stock_name} using {best_model_name}:")
        print(forecast_df.head())

        self.forecasts[stock_name] = forecast_df

        return forecast_df

    def test_cointegration(self, stock1, stock2, significance=0.05):
        """
        Test for cointegration between two stock series.

        Parameters:
        ----------
        stock1 : str
            Name of the first stock
        stock2 : str
            Name of the second stock
        significance : float, optional
            Significance level for the test

        Returns:
        -------
        tuple
            (is_cointegrated, coint_result)
        """
        if stock1 not in self.data or stock2 not in self.data:
            raise ValueError("Both stocks must be loaded before testing for cointegration")

        series1 = self.data[stock1]['Close']
        series2 = self.data[stock2]['Close']

        # Ensure series have the same length
        min_length = min(len(series1), len(series2))
        series1 = series1.iloc[-min_length:]
        series2 = series2.iloc[-min_length:]

        # Perform cointegration test
        coint_result = coint(series1, series2)
        p_value = coint_result[1]

        is_cointegrated = p_value < significance

        print(f"\nCointegration Test Results for {stock1} and {stock2}:")
        print(f"p-value: {p_value:.4f}")
        print(f"Result: {'Cointegrated' if is_cointegrated else 'Not cointegrated'}")

        if is_cointegrated:
            print("These stocks show a long-term equilibrium relationship, making them good candidates for pair trading.")

        return is_cointegrated, coint_result

    def plot_original_vs_differenced(self, stock_name):
        """
        Plot the original and differenced series.

        Parameters:
        ----------
        stock_name : str
            Name of the stock
        """
        if stock_name not in self.data:
            raise ValueError(f"No data found for {stock_name}")

        if stock_name not in self.stationary_data:
            self.difference_series(stock_name)

        original_series = self.data[stock_name]['Close']
        diff_series = self.stationary_data[stock_name]
        diff_order = self.differencing_order[stock_name]

        fig, axes = plt.subplots(2, 1, figsize=(12, 8))

        # Plot original series
        original_series.plot(ax=axes[0], title=f"{stock_name} - Original Series")
        axes[0].set_ylabel('Price')
        axes[0].grid(True)

        # Plot differenced series
        diff_series.plot(ax=axes[1], title=f"{stock_name} - {'Differenced' if diff_order > 0 else 'Original'} Series (d={diff_order})")
        axes[1].set_ylabel(f"{'Diff ' + str(diff_order) if diff_order > 0 else 'Price'}")
        axes[1].grid(True)

        plt.tight_layout()
        plt.show()

        # Also plot ACF and PACF of differenced series
        fig, axes = plt.subplots(2, 1, figsize=(12, 8))

        plot_acf(diff_series, ax=axes[0], lags=30)
        axes[0].set_title(f"ACF of {'Differenced' if diff_order > 0 else 'Original'} Series (d={diff_order})")

        plot_pacf(diff_series, ax=axes[1], lags=30)
        axes[1].set_title(f"PACF of {'Differenced' if diff_order > 0 else 'Original'} Series (d={diff_order})")

        plt.tight_layout()
        plt.show()

    def plot_residuals(self, stock_name):
        """
        Plot the residuals of the best fitted model.

        Parameters:
        ----------
        stock_name : str
            Name of the stock
        """
        if stock_name not in self.models:
            self.fit_models(stock_name)

        best_model_name = self.best_models[stock_name]
        model_result = self.models[stock_name][best_model_name]

        # Get residuals
        residuals = model_result.resid

        # Create figure with 3 subplots
        fig, axes = plt.subplots(3, 1, figsize=(12, 12))

        # Plot residuals
        axes[0].plot(residuals)
        axes[0].set_title(f"{stock_name} - Residuals of {best_model_name} Model")
        axes[0].grid(True)

        # Plot histogram of residuals
        axes[1].hist(residuals, bins=30, density=True, alpha=0.6)
        axes[1].set_title("Histogram of Residuals")

        # Add a normal density curve for comparison
        xmin, xmax = axes[1].get_xlim()
        x = np.linspace(xmin, xmax, 100)
        p = np.exp(-0.5 * ((x - residuals.mean()) / residuals.std())**2) / (residuals.std() * np.sqrt(2 * np.pi))
        axes[1].plot(x, p, 'k', linewidth=2)

        # Plot ACF of residuals
        plot_acf(residuals, ax=axes[2], lags=30)
        axes[2].set_title("ACF of Residuals")

        plt.tight_layout()
        plt.show()

        # Perform Ljung-Box test for autocorrelation
        lb_test = sm.stats.acorr_ljungbox(residuals, lags=[10, 20, 30], return_df=True)
        print(f"\nLjung-Box Test for Autocorrelation in Residuals:")
        print(lb_test)

        if (lb_test['lb_pvalue'] > 0.05).all():
            print("Residuals appear to be white noise (no significant autocorrelation).")
        else:
            print("Residuals show signs of autocorrelation, which may indicate model misspecification.")

    def plot_forecast(self, stock_name, include_history=True):
        """
        Plot the forecast with confidence intervals.

        Parameters:
        ----------
        stock_name : str
            Name of the stock
        include_history : bool, optional
            Whether to include historical data in the plot
        """
        if stock_name not in self.forecasts:
            self.forecast(stock_name)

        forecast_df = self.forecasts[stock_name]
        best_model_name = self.best_models[stock_name]

        plt.figure(figsize=(12, 6))

        if include_history:
            # Plot historical data
            history = self.data[stock_name]['Close']
            plt.plot(history.index, history, label='Historical')

            # Try to connect the last historical point to the first forecast point
            if isinstance(history.index[-1], pd.Timestamp) and isinstance(forecast_df.index[0], pd.Timestamp):
                plt.plot([history.index[-1], forecast_df.index[0]],
                         [history.iloc[-1], forecast_df['Forecast'].iloc[0]],
                         'k--', alpha=0.3)

        # Plot forecast
        plt.plot(forecast_df.index, forecast_df['Forecast'], 'r-', label='Forecast')
        plt.fill_between(forecast_df.index,
                         forecast_df['Lower CI'],
                         forecast_df['Upper CI'],
                         color='r', alpha=0.2, label='Confidence Interval')

        plt.title(f"{stock_name} - Forecast with {best_model_name} Model")
        plt.legend()
        plt.grid(True)
        plt.tight_layout()
        plt.show()

    def analyze_stock(self, stock_name, periods=30, plot=True):
        """
        Perform a complete analysis of a stock.

        Parameters:
        ----------
        stock_name : str
            Name of the stock
        periods : int, optional
            Number of periods to forecast
        plot : bool, optional
            Whether to generate plots

        Returns:
        -------
        dict
            Dictionary with analysis results
        """
        print(f"\n{'=' * 50}")
        print(f"Complete Analysis for {stock_name}")
        print(f"{'=' * 50}")

        # Check stationarity
        is_stationary, _, _ = self.check_stationarity(stock_name)

        # Difference if necessary
        if not is_stationary:
            self.difference_series(stock_name)

        # Fit models
        self.fit_models(stock_name)

        # Generate forecast
        forecast_df = self.forecast(stock_name, periods=periods)

        # Generate plots
        if plot:
            self.plot_original_vs_differenced(stock_name)
            self.plot_residuals(stock_name)
            self.plot_forecast(stock_name)

        return {
            'is_stationary': is_stationary,
            'differencing_order': self.differencing_order.get(stock_name, 0),
            'best_model': self.best_models[stock_name],
            'forecast': forecast_df
        }

    def analyze_multiple_stocks(self, stock_names, test_cointegration=True, periods=30, plot=True):
        """
        Perform a complete analysis of multiple stocks.

        Parameters:
        ----------
        stock_names : list
            List of stock names
        test_cointegration : bool, optional
            Whether to test for cointegration between pairs of stocks
        periods : int, optional
            Number of periods to forecast
        plot : bool, optional
            Whether to generate plots

        Returns:
        -------
        dict
            Dictionary with analysis results for each stock
        """
        results = {}

        for stock_name in stock_names:
            results[stock_name] = self.analyze_stock(stock_name, periods=periods, plot=plot)

        # Test for cointegration between pairs of stocks
        if test_cointegration and len(stock_names) > 1:
            print(f"\n{'=' * 50}")
            print("Cointegration Tests")
            print(f"{'=' * 50}")

            cointegration_results = {}

            for i, stock1 in enumerate(stock_names):
                for stock2 in stock_names[i+1:]:
                    is_cointegrated, coint_result = self.test_cointegration(stock1, stock2)
                    cointegration_results[(stock1, stock2)] = {
                        'is_cointegrated': is_cointegrated,
                        'p_value': coint_result[1]
                    }

            results['cointegration'] = cointegration_results

        return results

    def generate_investment_report(self, stock_names):
        """
        Generate an investment report based on the analysis.

        Parameters:
        ----------
        stock_names : list
            List of stock names

        Returns:
        -------
        str
            HTML report
        """
        # Ensure all stocks have been analyzed
        for stock_name in stock_names:
            if stock_name not in self.forecasts:
                self.analyze_stock(stock_name, plot=False)

        # Generate the report
        report = "<h1>Investment Analysis Report</h1>"

        # Summary of each stock
        report += "<h2>Individual Stock Analysis</h2>"

        for stock_name in stock_names:
            best_model = self.best_models[stock_name]
            forecast = self.forecasts[stock_name]

            # Calculate expected return (last forecast vs current price)
            current_price = self.data[stock_name]['Close'].iloc[-1]
            forecasted_price = forecast['Forecast'].iloc[-1]
            expected_return = (forecasted_price - current_price) / current_price * 100

            report += f"<h3>{stock_name}</h3>"
            report += f"<p>Current Price: ${current_price:.2f}</p>"
            report += f"<p>Best Model: {best_model}</p>"
            report += f"<p>Forecasted Price (30 periods ahead): ${forecasted_price:.2f}</p>"
            report += f"<p>Expected Return: {expected_return:.2f}%</p>"

            # Recommendation
            if expected_return > 5:
                recommendation = "Strong Buy"
            elif expected_return > 2:
                recommendation = "Buy"
            elif expected_return > -2:
                recommendation = "Hold"
            elif expected_return > -5:
                recommendation = "Sell"
            else:
                recommendation = "Strong Sell"

            report += f"<p><strong>Recommendation: {recommendation}</strong></p>"

        # Cointegration analysis if applicable
        if len(stock_names) > 1 and hasattr(self, 'cointegration_results'):
            report += "<h2>Pair Trading Opportunities</h2>"

            for (stock1, stock2), result in self.cointegration_results.items():
                if result['is_cointegrated']:
                    report += f"<p>{stock1} and {stock2} are cointegrated, suggesting a potential pair trading opportunity.</p>"

        # Final portfolio recommendation
        report += "<h2>Portfolio Recommendation</h2>"

        # Sort stocks by expected return
        stock_returns = []
        for stock_name in stock_names:
            current_price = self.data[stock_name]['Close'].iloc[-1]
            forecasted_price = self.forecasts[stock_name]['Forecast'].iloc[-1]
            expected_return = (forecasted_price - current_price) / current_price * 100
            stock_returns.append((stock_name, expected_return))

        stock_returns.sort(key=lambda x: x[1], reverse=True)

        report += "<p>Recommended allocation based on expected returns:</p>"
        report += "<ul>"
        for stock_name, expected_return in stock_returns:
            report += f"<li>{stock_name}: {expected_return:.2f}%</li>"
        report += "</ul>"

        return report


# Example usage
def main():
    # Create analyzer
    analyzer = TimeSeriesAnalyzer()

    # Get list of Excel files in the current directory
    import glob
    excel_files = glob.glob('*.xlsx')

    if not excel_files:
        print("No Excel files found in the current directory.")
        print("Please upload your Excel files and run the script again.")
        return

    print(f"Found {len(excel_files)} Excel files: {excel_files}")

    # Load data from each file
    stock_names = []
    for file in excel_files:
        stock_name = os.path.splitext(os.path.basename(file))[0]
        analyzer.load_data(file, stock_name)
        stock_names.append(stock_name)

    # Analyze all stocks
    results = analyzer.analyze_multiple_stocks(stock_names)

    # Generate investment report
    report = analyzer.generate_investment_report(stock_names)
    display(HTML(report))

if __name__ == "__main__":
    main()
```
